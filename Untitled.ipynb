```python
import re
import numpy as np # for data manipulation
import pandas as pd # for loading and data manipulation
import matplotlib.pyplot as plt # for drawing visualizations
import seaborn as sns # for drawing visualizations
```

```python
df = pd.read_csv(# load the dataset as dataframe, set tweetid as index column
    './data/train.csv', index_col='tweetid', encoding='utf-8'
) 
```

```python
# Let's creates a visual to remove the clean and split our tweets message to words
def make_tweet_words(tweet):
    # remove the old style retweet text "RT"
    tweet = re.sub(r'^RT[\s]+', '', tweet)
    # remove hyperlinks
    tweet = re.sub(r'http[s]?://(?:[A-Za-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9A-Fa-f][0-9A-Fa-f]))+', r'https', tweet.lower())
    # return a list of words
    return tweet.split(' ')

words_df = df.copy() # make a copy of our dataframe
words_df['message'] = words_df['message'].apply(make_tweet_words) # apply the make_tweet_words functionto create a list of words
words_df = words_df.explode('message').reset_index(drop=True) # unstack the list to rows and reset the index. drop current
print(f'The shape of the words dataframe is {words_df.shape}') # let's see the shape of our dataframe
words_df.head() # let's view the head of our dataframe
```

```python
hashtags = words_df[words_df['message'].str.startswith('#')] # filter hashtags from the words dataframe
print(f'The shape of the hastags dataframe is {hashtags.shape}') # let's see the shape of our dataframe
hashtags.head() # let's view the head of our dataframe
```

```python
handles = words_df[words_df['message'].str.startswith('@')] # filter twitter handles from the words dataframe
handles['message'] = handles['message'].apply(lambda word: word.strip(':')) # strip the ':' from the twitter handles
print(f'The shape of the twitter handles dataframe is {handles.shape}') # let's see the shape of our dataframe
handles.head() # let's view the head of our dataframe
```
